```python
import pandas as pd
```

```python
vine_df = pd.read_csv("vine_table.csv")
vine_df.head()
```

```python
high_rated_reviews_df = vine_df[(vine_df['total_votes'] >= 20)]
high_rated_reviews_df.head()
```

```python
helpful_reviews_df = high_rated_reviews_df
helpful_reviews_df['review_rating'] = high_rated_reviews_df['helpful_votes']/high_rated_reviews_df['total_votes']
helpful_reviews_df.head()
```

```python
good_helpful_reviews_df = helpful_reviews_df[(helpful_reviews_df['review_rating'] >= .5)]
good_helpful_reviews_df.head()
```

```python
paid_reviews_df = good_helpful_reviews_df[(good_helpful_reviews_df['vine'] == 'Y')]
paid_reviews_df
```

```python
unpaid_reviews_df = good_helpful_reviews_df[(good_helpful_reviews_df['vine'] == 'N')]
unpaid_reviews_df
```

```python
paid_reviews = paid_reviews_df['review_id'].count()
unpaid_reviews = unpaid_reviews_df['review_id'].count()
print(f"Paid Reviews: {paid_reviews}")
print(f"UnPaid Reviews: {unpaid_reviews}")
```

```python
paid_5_star = paid_reviews_df.loc[paid_reviews_df['star_rating'] == 5].count()
paid_5_star_count = paid_5_star['star_rating']
paid_5_star_count
```

```python
unpaid_5_star = unpaid_reviews_df.loc[unpaid_reviews_df['star_rating'] == 5].count()
unpaid_5_star_count = unpaid_5_star['star_rating']
unpaid_5_star_count
```

```python
percent_paid_5_star = paid_5_star_count/paid_reviews
percent_unpaid_5_star = unpaid_5_star_count/unpaid_reviews
print(f"Paid Reviews 5-star %: {percent_paid_5_star}")
print(f"UnPaid Reviews 5-Star %: {percent_unpaid_5_star}")
```
